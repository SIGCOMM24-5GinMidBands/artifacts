```python
def change_width(ax, new_value):
    for i, patch in enumerate(ax.patches):
        c_width = patch.get_width()
        diff = c_width - new_value
        patch.set_width(new_value)
        patch.set_x(patch.get_x() + diff * 0.5)
        if i == 0:
            patch.set_hatch(HATCHES['US3'])
            patch.set_edgecolor(colors['US3'])
            patch.set_facecolor('none')
        elif i == 1:
            patch.set_hatch(HATCHES['US2'])
            patch.set_edgecolor(colors['US2'])
            patch.set_facecolor('none')
        elif i == 2 or i == 3 or i == 4:
            patch.set_hatch(HATCHES['GER2'])
            patch.set_edgecolor(colors['GER2'])
            patch.set_facecolor('none')
        elif i == 5 or i == 6 or i == 7 or i == 8:
            patch.set_hatch(HATCHES['GER1'])
            patch.set_edgecolor(colors['GER1'])
            patch.set_facecolor('none')
        else:# i == 5 or i == 6 or i == 7 or i == 8:
            patch.set_hatch(HATCHES['US1'])
            patch.set_edgecolor(colors['US1'])
            patch.set_facecolor('none')
```

## Plotting Data

```python
fig = plt.figure(figsize=(5, 3.5))
gs1 = gridspec.GridSpec(1, 1, wspace=0.8, hspace=0.2, top=.85, bottom=0.3, left=0.12, right=0.96, figure=fig)


ax0 = plt.subplot(gs1[0])
ax0.yaxis.grid(True, which='major')


# tmpCDF = tmpCDF.dropna()
ax0 = sns.barplot(x='bw', y=y, data=data, estimator=np.max, ci=None)
change_width(ax0, 0.63)

ax0.set_ylim(-10, 350)
ax0.set_xlim(-0.5, 10.5)
ax0.set_yticklabels([0, 106, 162, 217, 245, 273], fontsize=12)
ax0.set_yticks([0, 106, 162, 217, 245, 273])#, fontsize=13)
ax0.set_ylabel('Resource Blocks ($N_{RB}$)', fontsize=12)
ax0.set_xticklabels(['40', '60', '80', '80', '80', '90', '90',
                        '90', '90', '100', '100'], fontsize=10)
ax0.set_yticks([i for i in range(0, 302, 100)])
ax0.set_yticklabels([t for t in range(0, 302, 100)], fontsize=12,)

ax0.set_xlabel('Channel bandwidth (in MHz)', fontsize=12)

cfsORder = ['Att_US', 'Vzw_US', 'S_Fr', 'V_It', 'V_Ge', 'O_Sp', 'V_Sp',
                        'O_Fr', 'T_Ge', 'Tmb_US', 'O_Sp']
for x in range(len(cfsORder)):
    text = ax0.text(x-0.1, 300, cfsORder[x], ha='center', va='center',
                            color='#800000', 
                    fontsize=10, zorder=2, rotation=90)


#### Showing plots
plot_name = 'Fig1'
plotme(plt, " Sec4", plot_name, show_flag=SHOW_PLOT_FLAG)
plt.close(fig)
```
