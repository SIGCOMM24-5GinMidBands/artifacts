## Loading Data (Per seconds)

```python
data = pd.read_csv('CuratedData/EU_Phy_DL_Spain_cqi_more.txt', sep='\t')

y = 'Tput(Mbps)'
```

## Plotting Data

```python
def change_width_def(ax, new_value):
#     for i, patch in ax.patches:
    for i, patch in enumerate(ax.patches):
#         print(i, patch.get_x(), patch.get_y(), str('{0:.2f}'.format(patch.get_height())))
        ax.text(patch.get_x() + 0.4, 950, str('{0:.1f}'.format(patch.get_height())),
                       ha='center', va='center', color='#800000', fontsize=10, zorder=20)
        c_width = patch.get_width()
        diff = c_width - new_value
        
#         ax.errorbar(patch.get_x() + 0.4, patch.get_height(), yerr=np.std(it1[x], ddof=1))
        
        patch.set_width(new_value)
        patch.set_x(patch.get_x() + diff * 0.5)
        if i == 1:
            patch.set_hatch(HATCHES['SPA1_1'])
            patch.set_edgecolor(colors['GER1'])
            patch.set_facecolor('none')
        elif i == 2:
            patch.set_hatch(HATCHES['SPA1_1'])
            patch.set_edgecolor(colors['US1'])
            patch.set_facecolor('none')
        elif i == 0:
            patch.set_hatch(HATCHES['SPA1_2'])
            patch.set_edgecolor(colors['ITL1'])
            patch.set_facecolor('none')
```

```python
fig = plt.figure(figsize=(2, 3.7))
gs1 = gridspec.GridSpec(1, 1, wspace=0.8, hspace=0.2, top=.85, bottom=0.3, left=0.12, right=0.96, figure=fig)
ax0 = plt.subplot(gs1[0])

order = [ 'V_Sp', 'O_Sp_100Mhz', 'O_Sp_90Mhz']
ax0.yaxis.grid(True, which='major')
sns.barplot(data=data, x='Config', y=y, ax=ax0, order=order,
            capsize=0.13, ci='sd', errwidth=1.7, alpha=0.85, estimator=np.median)
change_width_def(ax0, 0.5)


ax0.set_ylim(180, 1220)
ax0.set_yticks([i for i in range(0, 1220, 300)])
ax0.set_yticklabels([t for t in range(0, 1220, 300)], fontsize=12,)
ax0.set_ylabel('PHY DL Throughput\n(in Mbps)', fontsize=12,)
ax0.set_xlabel('Channel Bandwidth\n(in MHz)', fontsize=12,)

ax0.set_xlim([-0.4, 2.5])
ax0.yaxis.set_label_coords(-0.3, 0.4)
ax0.xaxis.set_label_coords(0.4, -0.18)
bw = [str(90), str(90), str(100)]
cfsORder = ['$V_{Sp}$', '$O_{Sp}$', '$O_{Sp}$' ]
for x in range(len(cfsORder)):
    text = ax0.text(x-0.02, 1110, cfsORder[x], ha='center', va='center',
                            color='#800000', 
                    fontsize=11, zorder=20, )

ax0.set_xticklabels(bw, fontsize=12,)



#### Showing plots
plot_name = 'Fig2'
plotme(plt, " Sec4", plot_name, show_flag=SHOW_PLOT_FLAG)
plt.close(fig)
```
