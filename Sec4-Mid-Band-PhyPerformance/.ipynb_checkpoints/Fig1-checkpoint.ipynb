# Figure 1

## Loading Data (Per seconds)

```python
data = pd.read_csv('CuratedData/EU_Phy_DL_All.txt', sep='\t')
us_data = pd.read_csv('CuratedData/US_Phy_DL_All.txt', sep='\t')
y = 'Tput(Mbps)'
```

## Plotting Data

```python
def add_median_labels(ax, precision='.1f', flag='noCA'):
    lines = ax.get_lines()
    boxes = [c for c in ax.get_children() if type(c).__name__ == 'PathPatch']
    lines_per_box = int(len(lines) / len(boxes))
    level = 1110
    l = 2.25
    for median in lines[4:len(lines):lines_per_box]:
        x, y = (data.mean() for data in median.get_data())
        value = x if (median.get_xdata()[1] - median.get_xdata()[0]) == 0 else y
        if flag != 'noCA':
            if x%2 == 0:
                l = l+0.1
            else:
                l = l-0.1
            text = ax.text(x, y + (l-y), f'{value:{precision}}', ha='center', va='center',
                            color='#800000',  fontsize=12, zorder=20)
            l = 2.25
        else:
            if x%2 == 0:
                level = level-20
            else:
                level = level+40
            text = ax.text(x, y + (level-y), f'{value:{precision}}', ha='center', va='center',
                            color='#800000',  fontsize=12, zorder=20)
            level = 1090
```

```python
fig = plt.figure(figsize=(6.3, 3))
gs1 = gridspec.GridSpec(1, 10, wspace=2.5, hspace=0.2, top=.85, bottom=0.3, left=0.12, right=0.96, figure=fig)
ax0 = plt.subplot(gs1[:6])

order = ['V_It',  'V_Sp', 'O_Sp_100Mhz', 'T_Ge', 'O_Fr','O_Sp_90Mhz']
ax0.yaxis.grid(True, which='major')
g = sns.boxplot(x='Config', y=y, data=data , dodge=True, width=0.5, showfliers=False, 
                order=order,  fliersize=0, ax=ax0)
add_median_labels(ax0)
ax0.set_ylim(180, 1220)
ax0.set_yticks([i for i in range(200, 1220, 200)])
ax0.set_yticklabels([t for t in range(200, 1220, 200)], fontsize=12)
ax0.set_ylabel('PHY Downlink\nThroughput (in Mbps)', fontsize=12)
ax0.set_xlabel('Europe Operators', fontsize=12)
bw = ['V_It',  'V_Sp', 'O_Sp', 'T_Ge', 'O_Fr','O_Sp' ]
ax0.set_xticklabels(bw, fontsize=12, rotation=25)


############ Plotting US data
ax1 = plt.subplot(gs1[7:10])
order_CA = ['US_Tmb', 'US_Vzw', 'US_Att']
sns.boxplot(x='Config', y='TP (Gbps)', data=us_data , dodge=True, width=0.5, showfliers=False, 
                order=order_CA,  fliersize=0, ax=ax1)
ax1.set_xlim([-0.5, 2.4])
ax1.set_ylim(-0.03, 2.53)
ax1.set_yticks([0, 0.5, 1, 1.5, 2, 2.5])
ax1.set_yticklabels([0, 0.5, 1, 1.5, 2, 2.5], fontsize=12)
ax1.set_ylabel('PHY Downlink\nThroughput (in Gbps)', fontsize=12)
ax1.set_xticklabels(['$Tmb_{US}$', '$Vzw_{US}$', '$Att_{US}$'], fontsize=12, rotation=25)
ax1.set_xlabel('U.S. Operators', fontsize=12)
ax1.yaxis.grid(True, which='major')
add_median_labels(ax1, flag='no')
#### Showing plots
plot_name = 'Fig1'
plotme(plt, " Sec4", plot_name, show_flag=SHOW_PLOT_FLAG)
plt.close(fig)
```
