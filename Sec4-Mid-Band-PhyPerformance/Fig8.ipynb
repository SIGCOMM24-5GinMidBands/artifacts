```python
import numpy as np
import pylab as pl
import matplotlib.pyplot as py

class Radar(object):

    def __init__(self, fig, titles, labels, rect=None):
        if rect is None:
            rect = [0.05, 0.05, 0.95, 0.95]

        self.n = len(titles)
        self.angles = [a if a <=360. else a - 360. for a in np.arange(90, 90+360, 360.0/self.n)]
        self.axes = [fig.add_axes(rect, projection="polar", label="axes%d" % i) 
                         for i in range(self.n)]

        self.ax = self.axes[0]
        self.ax.set_thetagrids(self.angles, labels=titles, fontsize=15, weight="bold", color="black")
        
        for ax in self.axes[1:]:
            ax.patch.set_visible(False)
            ax.grid(False)
            ax.xaxis.set_visible(False)
            self.ax.yaxis.grid(False)

        for ax, angle, label in zip(self.axes, self.angles, labels):
            ax.set_rgrids(range(1, 6), labels=label, angle=angle, fontsize=12, color='blue')
            ax.set_rlabel_position(ax.get_rlabel_position()+2)
            ax.spines["polar"].set_visible(False)
            ax.set_ylim(0, 6)  
            ax.xaxis.grid(True,color='black',linestyle='-')
        
        ax.annotate("PHY DL\n Tput.", xy=(0, 0), xycoords="data",
              xytext=(-4.71, 7), fontsize=16, color='#800000', weight='bold',
              va="center", ha="center",
              bbox=dict(boxstyle="round", fc="w", alpha=0),
         arrowprops=dict(arrowstyle="<-, head_length=0.9, head_width=0.3", color='#800000', lw=2.5))
        
    def plot(self, values, c, a, *args, **kw):
        angle = np.deg2rad(np.r_[self.angles, self.angles[0]])
        values = np.r_[values, values[0]]
        self.ax.plot(angle, values, color=c, *args, **kw)
        self.ax.fill(angle, values, color=c, alpha=a)
        self.ax.tick_params(axis='x', which='major', pad=10)

fig = pl.figure(figsize=(5, 5))

titles = [
" ", "# MIMO \nLayers  ", "Resource\nElements",  "Mode\nScheme", "    Channel\n         bandwidth",
]

labels = [
['  200', '', '  600', '', '  1Gbps'], ['  0', '', '  2x2', '', '  4x4'], 
    ['  0', '', '  2$x10^{10}$', '', '  4$x10^{10}$'],  ['  0', 'QPSK', '  16QAM', '64QAM', '  256QAM'],
    ['  20', '', '  60', '', '  100'],
]

radar = Radar(fig, titles, labels)
radar.plot([2.787, 3.45, 4.925,  4, 5,], colors['US1'],  0.3, linestyle='--',lw=3)
radar.plot([3.7985, 3.95, 3.75, 4.9, 4.5], colors['GER1'], 0.1, linestyle=':',  lw=3)
radar.plot([3.855, 3.86, 3.78, 4.7, 4.5], colors['ITL1'], 0.1, linestyle='-', lw=3)


h1, = plt.plot([-100, -100], linestyle='--', color=colors['US1'], lw=5)
h2, = plt.plot([-1000, -1000], linestyle=':', color=colors['GER1'], lw=5)
h3, = plt.plot([-1000, -1000], linestyle='-', color=colors['ITL1'], lw=5)
h4, = plt.plot([-1000, -1000], linestyle='-', color='yellow', lw=5)

legend1 = radar.ax.legend((h3, h2, h1), ('V_Sp [90]', 'O_Sp [90]',  'O_Sp [100]'),
                     columnspacing=1,
                     loc='center', ncol=3,
                    bbox_to_anchor=(0.55, 0.87), facecolor='#dddddd',
                     handlelength=3, framealpha=0.1, fontsize = 14)#, markerscale=40)
radar.ax.add_artist(legend1)


#### Showing plots
plot_name = 'Fig8'
plotme(plt, " Sec4", plot_name, show_flag=SHOW_PLOT_FLAG)
plt.close(fig)
```
